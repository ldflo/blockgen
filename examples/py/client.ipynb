# list_todos

```python
""" List all todos """
import requests

BASE_URL = "http://127.0.0.1:8000"

response = requests.get(f"{BASE_URL}/todos")
print(response.status_code)
print(response.text)
```

# create_todo

```python
""" Create a new todo """
import requests

BASE_URL = "http://127.0.0.1:8000"

body = {
    "title": "...", # Change me
    "done": False, # Change me
}
response = requests.post(f"{BASE_URL}/todos", json=body)
print(response.status_code)
print(response.text)
```
